# Binomial disribution converges to normal distribution when number of samples are large enough and probability of sucess and failure are not too small.

## Such that mean = n*p and standard deviation = n*p*(1-p)

```python
import numpy as np
import math
import scipy.special
import matplotlib.pyplot as plt
```

```python
#Binomial PDF
#P(x=k) = nCk (p)^k * (1-p)^(n-k)
```

Normal distribution PDF


```python
def normal_dist(x, mean, sd):
    prob_density = (1/(sd*math.sqrt(2*np.pi))) * np.exp((mean-x)**2/(2*sd**2))
    return prob_density

def binomial_dist(k):
    prob_density = scipy.special.binom(n, k) * p**k * q**(n-k)
    return prob_density
```

```python
n = 100 #increase n for better approximation
p = 0.5 #Should not be too small 
q = 1 - p #Should not be too small
```

```python
pdf_bi = []
pdf_nor = []
mu = n*p
sigma = math.sqrt(n*p*q)
for k in range(n):
    pdf_bi.append(k)
    pdf_nor.append(normal_dist(k,mu,sigma))

k = [i for i in range(n)]
plt.plot(k, pdf_bi, 'r')
plt.plot(k, pdf_nor, 'g')
plt.ylabel('P(X=k) pdf')
plt.xlabel('k values')
plt.legend(["PDF of Binomial","PDF of Normal"])
plt.show()
```
